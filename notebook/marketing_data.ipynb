# Data validasi

## Cek Shape data

```python
def check_shape_data(df,nama_tabel):
    rows_amount = df.shape[0]
    columns_amount = df.shape[1]

    print(f'Data {nama_tabel} has {rows_amount} rows and {columns_amount} columns')

check_shape_data(df,data_name)
```

## Cek tipe data

```python
def check_type_data(df, table_name):
    print(f'Each column in `{table_name}` has the following data types:\n')
    for column in df.columns:
        print(f'Column `{column}` has data type `{df[column].dtype}`')

check_type_data(df, data_name)
```

## Check missing value

```python
def check_missing_values(df, table_name):
    print(f'Each column in `{table_name}` has the following number of missing values:\n')
    for column in df.columns:
        missing_values = df[column].isnull().sum()
        percentage_missing_values = round(missing_values/len(df)*100,1)
        print(f'Column `{column}` has {missing_values} or {percentage_missing_values}% missing values')

check_missing_values(df, data_name)
```

## Check duplicate

```python
def check_duplicates(df, table_name):
    duplicate_count = len(df[df.duplicated()])
    print(f"`{table_name}` has {duplicate_count} duplicate entries")

check_duplicates(df, data_name)
```

## Check unique values

```python
def check_unique_values(df,table_name):
    print(f'Each column in `{table_name}` has the following number of missing values:\n')
    for column in df.columns:
        unique_values = df[column].unique()
        if len(unique_values)<=50:
            print(f'''Jumlah unique values pada kolom `{column} `adalah sebanyak : {len(unique_values)} yaitu : \n {unique_values}\n''')
        else:
            print(f'''Jumlah unique values pada kolom `{column}` adalah sebanyak : {len(unique_values)}\n''')

check_unique_values(df, data_name)
```

# Cleaning

## Handle Incosistent data Value

```python
import re
import numpy as np

def cleaning_unique_values_prices_availability(prices_availability):
    prices_availability = prices_availability.upper()
    yes_availabilty = ['YES','TRUE','SPECIAL ORDER','IN STOCK','32 AVAILABLE','7 AVAILABLE']
    no_availabilty = ['UNDEFINED','OUT OF STOCK','NO','MORE ON THE WAY','SOLD','FALSE','RETIRED']

    if prices_availability in yes_availabilty:
        return 'YES'
    elif prices_availability in no_availabilty :
        return 'NO'
    
    return prices_availability

def convert_to_pounds(weight_str):
    try:
        pounds = 0
        ounces = 0

        # Conversion factors
        ounces_to_pounds = 1 / 16
    
        
        # Find pounds or lbs (only take the first occurrence)
        pounds_match = re.search(r'(\d*\.?\d+)\s?(?:lbs?|pounds?)', weight_str)
        if pounds_match:
            pounds = float(pounds_match.group(1))
        
        # Find ounces values
        ounces_matches = re.search(r'(\d*\.?\d+)\s?(?:oz|ounces?)', weight_str)
        if ounces_matches:
            ounces = float(ounces_matches.group(1))
        
        # Convert ounces to pounds and sum
        total_pounds = pounds + (ounces * ounces_to_pounds)
        return total_pounds
    except:
        return np.NaN

## Handle Incosistent data Value
# Apply each function to the column
df['weightInPounds'] = df['weight'].apply(convert_to_pounds)
df['prices.availability'] = df['prices.availability'].apply(cleaning_unique_values_prices_availability)
df['prices.condition'] = df['prices.condition'].str.upper().str.replace('NEW OTHER (SEE DETAILS)','NEW')
df = df[~(df['prices.condition'].isin(list(df['prices.condition'].unique()[-2:])))]
df['prices.shipping'] = df['prices.shipping'].fillna('free').apply(lambda x: 0 if 'USD' not in x else x.replace('USD','').replace('.','')).astype('int')
df = df.rename(columns={'id':'product_id'})
# Drop unknown and 'weight' columns
for col in df.columns:
    if 'UNNAMED' in col.upper():
        df = df.drop(col, axis=1)
    if col.upper()=='WEIGHT':
        df = df.drop(col,axis=1)

## Handling missing values
# fill missing value in column manufacturer
df['manufacturer'] = df['manufacturer'].fillna('No Information')
# Mengubah kolom menjadi datetime
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])

# Drop because missing value > 75%
df = df.drop('ean',axis=1)
```

## Rename the columns

```python
# Rename columns if "." in column , change to "_"
rename_columns = {}

for column in df.columns:
    if '.' in column:
        rename_columns[column] = '_'.join(column.split('.'))

df = df.rename(columns=rename_columns)

# Ubah nama kolom dari camelCase ke snake_case
# Fungsi untuk mengubah camelCase ke snake_case
def camel_to_snake(column_name):
    # Ganti 'URLs' dengan 'urls'
    column_name = re.sub(r'RLs', 'rls', column_name)
    # Konversi camelCase ke snake_case
    return re.sub(r'(?<!^)(?=[A-Z])', '_', column_name).lower()

# Mengubah nama kolom dengan rename() menggunakan fungsi camel_to_snake
df = df.rename(columns=camel_to_snake)
df.columns
```
